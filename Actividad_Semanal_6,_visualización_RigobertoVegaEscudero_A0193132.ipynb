Las medidas de tendencia central son medidas estadísticas que pretenden resumir en un solo valor a un conjunto de valores. Representan un centro en torno al cual se encuentra ubicado el conjunto de los datos. Las medidas de tendencia central más utilizadas son: media, mediana y moda. Las medidas de dispersión en cambio miden el grado de dispersión de los valores de la variable. Dicho en otros términos las medidas de dispersión pretenden evaluar en qué medida los datos difieren entre sí. De esta forma, ambos tipos de medidas usadas en conjunto permiten describir un conjunto de datos entregando información acerca de su posición y su dispersión.

Aquí se adjunta la descripción de las medidas de tendencia central y dispersión

count: Es la cantidad de datos sin NA por columna.

mean: Es la suma de todos los datos, dividido entre la cantidad de datos, todos distintos a NA.

std: Desviación estándar, da una idea de qué tan dispersos o alejados se encuentran los datos del conjunto respecto del mean de cada columna.

min: Es el valor mínimo de cada columna.

25%: Es el primer cuartil. Se ordena el conjunto de menor a mayor, este es el valor que se encuentra en el lugar (0.25 * count) de cada columna.

50%: Este es el valor que se encuentra en el lugar (count * 0.50) de cada columna.

75%: Este es el valor que se encuentra en el lugar (count * 0.75) de cada columna.

max: Es el valor máximo de cada columna.

#6. Escala los datos, si consideras necesario

La estandarización de datos, en ocasiones también conocida como normalización, es el proceso de ajustar o adaptar ciertas características para que los datos se asemejen a un tipo, modelo o normal común con el objetivo de que su tratamiento, acceso y uso sea más sencillo para los usuarios o personas que dispongan de ellos.

El propio concepto de datos abiertos implica una búsqueda de la estandarización en el uso de la información en formato abierto para que pueda utilizarse y ser reutilizada por parte de la ciudadanía.

Por tanto, podría decirse que la estandarización es la forma en la que todas las personas pueden comparar y consultar datos y encontrar siempre la información que necesitan, teniendo certeza de uniformidad en la forma en la que lo encontrarán.

Normalización de puntuación Z (Estandarización) - convierte los datos del conjunto a tener una media de 0 y una desviacion estandar de 1 

 $$ x_i^{new} = \cfrac{x_i^{old} - \mu}{\sigma}$$ 

 La fórmula se define como  la resta entre el valor x en turno de nuestro arreglo menos el promedio de los valores del arreglo x, todo entre la desviación etandar del arreglo x. ((X-xm)/sd)

 Cabe resaltar que la estandarización es de los metodos más utilizados para hacer PCAs.

```python
#Extraer Variables Numerica
numericas = ['X1', 'X5', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23']
df_main_num = df_main[numericas]
df_main_num
```

```python
#Definimos una función de estandarización o normalización.

def standarization(X):
  xm = np.mean(X)
#El metodo statisctics.pstdev(), calcula la desviación estandar de un arreglo
  sd =sts.pstdev(X)
  return ((X-xm)/sd)
```

```python
#A continuación normalizamos  todas las columnas en el rango de 0-1, usando la función previamente definida normalize
#esta función re escala la información de valores entre 0 y 1
#Nuestro nuevo dataframe de valores normalizados se llama´ra df_2 normalizew
columnas = df_main_num.columns
print(columnas)
df_main_estandarizado = df_main_num.copy()
for n in columnas:
  df_main_estandarizado[n] = standarization(df_main_num[n])
df_main_estandarizado
```

#7. Reduce las dimensiones con PCA, si consideras necesario.
1. Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
1. Indica la importancia de las variables en cada componente

```python
#Generamos un nuevo dataset llamado varianza, el cual contendra la covarianza cov del data set previamente normalizado
#Considerando que el dataset se traspone con el .T, por lo que toma una forma mas simple de analisis en cuanto cantidad de columnas y filas

matriz_cov = df_main_estandarizado.corr()
```

```python
#Generamos un Heatmap para la matriz de correlacion llamada varianza
sns.set(rc={'figure.figsize':(20,20)}) 
sns.heatmap(matriz_cov, cmap = 'YlGnBu', annot=True,)
```

```python
#Procedmeos a hacer un estudio PCA de 5 componentes minimos a ver si cumplimos con el 70% de la varianza total
componente_minimos = 5
pca = PCA(n_components=componente_minimos)
pca_data = pca.fit_transform(df_main_estandarizado)


print('Varainza de cada componente minimo\n')
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pca.explained_variance_),
                             'Proportion of variance': pca.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pca.explained_variance_ratio_)
                             })
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df
#Sacamos valores estadisticos de los pesos.
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)

```

Podemos ver que cumplimos con la varianza pedida del 70%

Ahora podemos notar que con 5 componentes mínimos pudimos llegar al objetivo de tener un mínimo de 70% de la varianza total.

```python
#sacamos varianzas de acuerdo a los pesos
pcsComponents_df = pd.DataFrame(pca.components_.transpose(), 
                                columns = pcsSummary_df.columns,
                                index = df_main_estandarizado.columns
                                )
pcsComponents_df
```


Tomando en cuenta que estamos usando el dataset limpio y que se redujo todo a solo 5 columnas de datos, valores que representan mayores pesos en estas varianzas de acuerdo a la tabla son principales pesos serian las siguientes:

```python
names = pcsComponents_df.columns
x = 1
for n in names:
 print(f"Variable con mas peso en PC{x}:", pcsComponents_df[n].idxmax(axis=0, skipna=False))
 x+=1
```

```python
PC_components = pcsSummary_df.columns
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pca.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components, 
                 y = np.cumsum(pca.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Histograma')
plt.xlabel('N-th Componentes principales')
plt.ylabel('Varianza explicada')
plt.ylim(0, 1)
plt.show()
```
