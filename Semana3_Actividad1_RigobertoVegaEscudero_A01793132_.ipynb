# **Parte 1: Fundamentos de bases de datos**
##1. Fundamentos de bases de datos y para ciencia de datos.  

La ciencia de datos es uno de los campos que más rápido están creciendo. Debido a que día a día la dependencia de los datos se va haciendo cada vez más grande, por ello se requiere de personas capacitadas para el conocimiento de los datos, obtenerlos, limpiarlos, analizarlos, visualizarlos y usarlos para construir soluciones.

Por ende, un científico de datos necesitara crear las bases de datos necesarias para dar soluciones en base a metodologías de científicos de datos requeridos en los proyectos específicos.

Por ello, y viendo que los datos nos envuelven día a día más, colectar y analizar los datos nos dicta el uso y comprensión de estas herramientas, para hacer estructuras sencillas, pero, a su vez eficientes, seguras y rápidas.

Una base datos está definida como una estructura de conjuntos de datos almacenados en una memoria de la computadora o en la nube que puede ser acusada de varias maneras. Esta información debe tener una lógica entre los datos almacenados, integrando registros autodescriptivos, para con ello modelo dar sistemas particulares de un mundo real.

Estos datos provienen de fuentes de información variadas y estas pueden ser almacenadas de distintas formas y estas caen en una vista que muestra esta información en una combinación de registros y campos, definidas como tablas. Otra forma de verlo es como una colección de datos relacionados a las tablas donde dichas tablas tienen relaciones entre los datos que van obteniendo, esto definiendo la estructura central de las bases de datos.

La recopilación de los datos debe ser compartida dando lógica entre las relaciones de los mismos, para con ello satisfacer necesidades organizacionales. Estos datos pueden tener entidades, atributos y relaciones de la información de una organización.

El enfoque de las bases de datos es básicamente, tener repositorios centrales, datos gestionados por un agente y almacén de forma estadística y conveniente al usuario.

Para ello se usan sistemas DBMS - Database Management Systems. Esto se define como un software que se utiliza para crear, mantener y proporcionar acceso controlado a los usuarios de una base de datos. Este programa interactúa con la base de datos mediante una solicitud ad hoc al DBMS.

Los dos lenguajes más conocidos:
1. DML, también conocido como lenguaje de consulta (Query)
2. SQL (Structured Query Language), siendo este el más utilizado.

Con SQL viene una herramienta poderosa para manipular los datos. Ya que es fácil de usar, poderosa y eficiente.

Gracias a este lenguaje las bases de datos se han dividido en dos tipos:

1. Bases de datos relacionables: estas son las que los datos están organizados y almacenados dentro de tablas que pueden ser conectadas a otras con alguna clase de relación.
2. Bases de datos no relacionables: Estas bases de datos también son llamadas NoSQL. Debido a que estas consiguen conectarse por medio de categorías más que relaciones entre las mismas.



###2.Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.
Un almacén de datos es básicamente una base de datos gigante que contiene información de diferentes fuentes y las empata juntas para su uso. Estas comúnmente están diseñadas en una relación normalizada de una base de datos, ya que estas tienden a contener datos históricos de accesos de datos del pasado con los actuales, tablas actualizadas y versiones pasadas que muchas veces son actualizadas en tiempo real, etc.

Un almacenamiento de datos es un repositorio central de información que se puede analizar para tomar decisiones mejor informadas. Los datos fluyen hacia un almacenamiento de datos desde sistemas transaccionales, bases de datos relacionales y otros orígenes, normalmente a una cadencia regular. 

Dentro de ello logra centralizar y consolidar toda esta Big Data. Dotando al usuario una manera de ejecutar queries y análisis de manera que tenga este contenedor central que lo ayude y por ende extraer información valiosa de los mismos.

La arquitectura de las bases de datos se basa en:

1. Simple: Todos los almacenes de datos comparten un diseño simple en donde la metadata, el resumen de los datos y el bruto de la data están almacenados dentro del repositorio central del almacén.
2. Simple con escenificación del área: Los datos operacionales deben ser limpiados y procesados antes de ser puestos en el almacén. Aunque esto puede ser ejecutado por algún programa, muchos almacenes de dados añaden un área de puesta de escena para los datos, antes que estos entren al almacén.
3. Centro y habla: Añadir mercados de datos entre los repositorios centrales y los usuarios finales permiten a la organización personalizar sus almacenes de datos para servir a las diferentes líneas del negocio.
4. Cajas de arena: Las cajas de arena son privadas y seguras. Estas áreas seguras permiten a las compañías explorar e informar rápidamente acerca de nuevos conjuntos de datos o maneras de analizar información sin tener que conformarse a solo compilarla dentro de las reglas y protocolos tradicionales de los almacenes de datos.

Dentro de los beneficios de un almacenamiento de datos se incluyen los siguientes:

- Toma de decisiones fundamentadas
- Datos consolidados provenientes de muchos orígenes
- Análisis de datos históricos
- Calidad, coherencia y precisión de los datos
- Separación del procesamiento de análisis de las bases de datos transaccionales, lo que mejora el rendimiento de ambos sistemas
